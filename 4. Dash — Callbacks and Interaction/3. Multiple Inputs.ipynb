# 2 Inputs 1 Output

This little notebook will be so cool that it will help you to check the relationship between two features while you are analyzing a dataset in the EDA process.

Here we will make a Graph which will be connected to 2 drop downs and then the points will be plotted accordingly. Well which will help to run any dataset to check the relationship.

— <br>
First, I will try it myself and then will check the lecture. If there is something change then will update or make a new code. So let's get started.

```python
import dash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Input, Output

import plotly.graph_objs as go
import plotly.offline as pyo

import pandas as pd
```

```python
df = pd.read_csv("../DATA/Data/mpg.csv")
df
```

### *My Way*

```python
app = dash.Dash()
app.layout = html.Div(children=[
                        html.Div([
                            dcc.Dropdown(id="filter-1", 
                                         options=[{"label":feature, "value":feature}
                                                  for feature in df.columns])], 
                            style=dict(width="50%", display="inline-block")
                        ),
                        html.Div([
                            dcc.Dropdown(id="filter-2", 
                                         options=[{"label":feature, "value":feature}
                                                  for feature in df.columns])],
                            style=dict(width="50%", display="inline-block")
                        ),
                        dcc.Graph(id="graph")
])

@app.callback(Output("graph", "figure"),
             [Input("filter-1", "value"), 
              Input("filter-2", "value")])

def update_graph(input_feature1, input_feature2):
    
    #  ↓ Only update if both inputs are NON None
    if all([input_feature1, input_feature2]): 
        traces = [
                    go.Scatter(x=df[input_feature1],
                               y=df[input_feature2],
                               mode="markers",
                               marker={"size":10, "opacity":0.5,
                                       "line":dict(width=0.5, color="white")})
                 ]
        layout = dict(title=f"{input_feature1.title()} vs {input_feature2.title()}",
                     xaxis=dict(title=f"← {input_feature1.title()} →"),
                     yaxis=dict(title=f"← {input_feature2.title()} →"))
        return {"data":traces, "layout":layout}


app.run_server()
```

Works! My code works!

Thigs to know:
- I have provided the Dropdowns in their separate Divs. So they are more managable and style will work good
- I have given the style - display = inline-block so both dropdowns are side by side
- Gave a condition in the update function to check for None which comes when the no values are selected.

# 4 Inputs 1 Output
*With Filters*

```python
app = dash.Dash()
app.layout = html.Div(children=[
                        html.Div([
                            html.Label("X-Axis"),
                            dcc.Dropdown(id="filter-1", 
                                         options=[{"label":feature, "value":feature}
                                                  for feature in df.columns])], 
                            style=dict(width="50%", display="inline-block")
                        ),
                        html.Div([
                            html.Label("Y-Axis"),
                            dcc.Dropdown(id="filter-2", 
                                         options=[{"label":feature, "value":feature}
                                                  for feature in df.columns])],
                            style=dict(width="50%", display="inline-block")
                        ),
                        dcc.Graph(id="graph"),
                        html.Div([
                            html.Label("Size"),
                            dcc.Dropdown(id="filter-3", 
                                         options=[{"label":feature, "value":feature}
                                                  for feature in df.columns])],
                            style=dict(width="50%", display="inline-block")
                        ),
                        html.Div([
                            html.Label("Color"),
                            dcc.Dropdown(id="filter-4", 
                                         options=[{"label":feature, "value":feature}
                                                  for feature in df.columns])],
                            style=dict(width="50%", display="inline-block")
                        ),
])

@app.callback(Output("graph", "figure"),
             [Input("filter-1", "value"), 
              Input("filter-2", "value"),
              Input("filter-3", "value"),
              Input("filter-4", "value")])

def update_graph(input_feature1, input_feature2, input_size, input_color):
    
    #  ↓ Only update if both inputs are NON None
    if all([input_feature1, input_feature2]): 
        if input_size:
            input_size = df[input_size]
        if input_color:
            input_color = df[input_color]
        
        traces = [
                    go.Scatter(x=df[input_feature1],
                               y=df[input_feature2],
                               mode="markers",
                               marker={"size":input_size, "opacity":0.5,
                                       "line":dict(width=0.5, 
                                                   color="white"),
                                      "color":input_color},
                               colorscale=True)
                 ]
        layout = dict(title=f"{input_feature1.title()} vs {input_feature2.title()}",
                     xaxis=dict(title=f"← {input_feature1.title()} →"),
                     yaxis=dict(title=f"← {input_feature2.title()} →"))
        return {"data":traces, "layout":layout}

app.run_server()
```
